```python
"""
BF550 Project 0
Code to play tic-tac-toe against computer, with variable board size
Kyrah Kotary
kkotary@bu.edu
"""

import random
from random import randint

board = [['', '', ''], ['', '', ''], ['', '', '']]  # The game board
win_board = [['', '', ''], ['', '', ''], ['', '', '']]  # the winning board

#variables 
X = 'X'
O = 'O'
player = 'Player'

#initialize normal 3 x 3 board with 3 in a row for win
#can be changed later
rows = 3
columns = 3
k = 3

def empty_cells(): #track which spaces are empty
    empty = []

    for i in range(columns):
        for j in range(rows):
            if board[i][j] != X and board[i][j] != O:
                empty.append((i, j))

    return empty


def get_user_move(p1, p2):
    
    x = int(input(f'please enter row number (1 - {rows}) :')) - 1 
    y = int(input(f'please enter column number (1 - {columns}):')) - 1

    if x >= rows or y >= columns or board[x][y] == p1 or board[x][y] == p2:
        print('Please Enter A Valid Move')
        get_user_move(p1, p2)
        return

    board[x][y] = p1
    print()
    

def AI_0(ai, pl, x=0): #chooses random space from avilable spaces
    
    available = []

    for x, y in empty_cells():
        if board[x][y] != X and board[x][y] != O:
            available.append((x, y))

    ai_move = random.choice(available)
    board[ai_move[0]][ai_move[1]] = ai 
    print('ai played ', 'row', (ai_move[0])+1, 'and column', (ai_move[1])+1)
    
    return

def AI_1(ai, pl, x=0): #looks for win or chooses random
  
    # Rows
    for i in range(rows):
        for j in range(columns):
            if all(board[i][j] == ai for j in range(k - 1)):
                if board[i][j] != X and board [i][j] != O:
                    for j in range(rows):
                        board[i][j] = ai
                        print('ai played ', 'row', (board[i])+1, 'and column', (board[j])+1)
                    return
                    break
            else:
                continue
            
    # Columns            
    for j in range(columns):
        for i in range(rows):
            if all(board[j][i] == ai for j in range(k - 1)):
                if board[j][i] != X and board [j][i] != O:
                    for j in range(columns):
                        board[j][i] = ai
                        print('ai played ', 'row', (board[i])+1, 'and column', (board[j])+1)
                    return
                    break
            else:
                continue
                
    #Diagonals            
    for i in range(rows):
        for j in range(columns):
            if all(board[i][i] == ai for i in range(k - 1)):
                if board[i][i] != X and board [i][i] != O:
                    for i in range(rows):
                        board[i][i] = ai
                        print('ai played ', 'row', (board[i])+1, 'and column', (board[j])+1)
                    return
                    break
            else:
                continue
    
    for i in range(rows):
        if all(board[i][-(i + 1)] == ai for i in range(k - 1)):
            if board[i][-(i + 1)] != X and board [i][-(i+1)] != O:
                for i in range(rows):
                    board[i][-(i+1)] = ai
                    print('ai played ', 'row', (board[i])+1, 'and column', (board[j])+1)
                return
                break
            else:
                continue
        
    
    available = []

    for x, y in empty_cells():
        if board[x][y] != X and board[x][y] != O:
            available.append((x, y))

    ai_move = random.choice(available)
    board[ai_move[0]][ai_move[1]] = ai 
    print('ai played ', 'row', (ai_move[0])+1, 'and column', (ai_move[1])+1)
    
    return
        
    
def AI_2(ai, pl, x=0): #blocks, looks for win, or chooses random

#blocking
    #Rows
    for i in range(rows - 1):
        for j in range(columns):
            if board[i][j] == pl and board[i+1][j] == pl:
                for i in range(rows):
                    if board[-i][j] != X and board [-i][j] != O:
                    #for i in range(rows):
                        board[-i][j] = ai
                        return
                        break
            else:
                continue
    #Columns            
    for i in range(rows):
        for j in range(columns - 1):
            if board[i][j] == pl and board[i][j+1] == pl:
                for j in range(columns):
                    if board[i][-j] != X and board [i][-j] != O:
                    #for j in range(columns):
                        board[i][-j] = ai
                        return
                        break
            else:
                continue
                
     #Diagonal          
    for i in range(rows - 1):
        #for j in range(columns - 1):
        if board[i][i] == pl and board[i+1][i+1] == pl:
            for i in range(rows):
                if board[i][i] != X and board [i][i] != O:
                    board[i][i] = ai
                    return
                    break
            else:
                continue
                   
 #winning                   
    # Rows
    for i in range(rows):
        for j in range(columns):
            if all(board[i][j] == ai for j in range(k - 1)):
                if board[i][j] != X and board [i][j] != O:
                    for j in range(rows):
                        board[i][j] = ai
                    return
                    break
            else:
                continue
            
    # Columns            
    for j in range(columns):
        for i in range(rows):
            if all(board[j][i] == ai for j in range(k - 1)):
                if board[j][i] != X and board [j][i] != O:
                    for j in range(columns):
                        board[j][i] = ai
                    return
                    break
            else:
                continue
                
    #Diagonals            
    for i in range(rows):
        for j in range(columns):
            if all(board[i][i] == ai for i in range(k - 1)):
                if board[i][i] != X and board [i][i] != O:
                    for i in range(rows):
                        board[i][i] = ai
                    return
                    break
            else:
                continue
    
    for i in range(rows):
        if all(board[i][-(i + 1)] == ai for i in range(k - 1)):
            if board[i][-(i + 1)] != X and board [i][-(i+1)] != O:
                for i in range(rows):
                    board[i][-(i+1)] = ai
                return
                break
            else:
                continue
        
  #random  
    available = []

    for x, y in empty_cells():
        if board[x][y] != X and board[x][y] != O:
            available.append((x, y))

    ai_move = random.choice(available)
    board[ai_move[0]][ai_move[1]] = ai 
    print('ai played ', 'row', (ai_move[0])+1, 'and column', (ai_move[1])+1)
    
    return

        
    
def get_win(p): #keep track of winning board

    for i in range(rows):
        # Rows
        if all(board[i][j] == p for j in range(k)):
            for j in range(k):
                win_board[i][j] = p
            return

        # Columns
        if all(board[j][i] == p for j in range(k)):
            for j in range(k):
                win_board[j][i] = p
            return

    #Diagonals
    if all(board[i][i] == p for i in range(k)):
        for i in range(k):
            win_board[i][i] = p

        return

    if all(board[i][-(i + 1)] == p for i in range(k)):
        for i in range(k):
            win_board[i][-(i + 1)] = p

        return


def display_board(): #print the game board

    for i in range(rows - 1):
        for j in range(columns - 1):
            print(board[i][j], end=' | ')
        print(board[i][-1])
        print('---' * columns + '-' * (columns - 3))

    for j in range(columns - 1):
        print(board[-1][j], end=' | ')

    print(board[-1][-1])
    print()


def display_win(p): #print the winning board

    get_win(p)

    for i in range(rows - 1):
        for j in range(columns - 1):
            print(win_board[i][j], end=' | ')
        print(win_board[i][-1])
        print('---' * columns + '-' * (columns - 2))

    for j in range(columns - 1):
        print(win_board[-1][j], end=' | ')

    print(win_board[-1][-1])
    print()



def change_board(): #change board parameters
    
    global rows
    global columns
    global k
    rows = int(input('Please Enter rows: '))
    columns = int(input('Please Enter columns: '))
    k = int(input('How many in a row constitutes a win?: '))
    initialize()


def start_game(): #starting game menu

    print('Welcome to Tic Tac Toe!')

    print( 'change board settings (1), play (2), or quit (3)')
    option = input('\nWhat would you like to do?: ')
        
    if option == '1':
        change_board()

    if option == '2':
        initialize()
        play_game('X', 'O')

    if option == '3':
        print('Goodbye!\n')
        return

    print()
    start_game()


def play_game(p1, p2): #the actual game

    print()
    initialize()

    computer = 'ai'

    print('1. Easy')
    print('2. Medium')
    print('3. Hard')

    print()

    level = int(input('Please choose a level to play against: '))

    print()

    while computer == player:
        computer = 'ai'


    #choose which player goes first
    c = randint(0, 1) 

    pl = p1
    ai = p2

    if c == 0:
        ai = p1
        pl = p2

        print('\n' + 'Computer goes First!\n\n')

    else:
        print('\n' + 'Player goes First!\n\n')
        
        display_board()

    turn = 0

    while True:
        t = turn % 2

        if t == c:
            if level == 1: AI_0(ai, pl, 1)
            if level == 2: AI_1(ai, pl, 1)
            if level == 3: AI_2(ai, pl, 1)

            display_board()

            if win(ai):
                print(computer + ' Wins!\n')
                print('Below Is How ' + 'computer' + ' Won\n\n')
                display_win(ai)
                break

        else:
            get_user_move(pl, ai)

            display_board()

            if win(pl):
                print(player + ' Wins!')
                print('Below Is How ' + 'player' + ' Won\n')
                display_win(pl)
                break
                
        if len(empty_cells()) == 0:
            print('Tie!')
            break

        turn += 1

    play_again(p1, p2)


def initialize(): #reset the board

    global board, win_board

    board = [[' ' for _ in range(rows)] for __ in range(columns)]
    win_board = [[' ' for _ in range(rows)] for __ in range(columns)]


def play_again(p1, p2): #asks user if they want to play agin

    option = input('Would You Like To Play Again? (y(yes) / n(no): ').lower()

    if option == 'y':
        play_game(p1, p2)
    elif option == 'n':
        return
    else:
        print('\nPlease Enter a Valid Option')
        play_again(p1, p2)


def win(p): #checks if either of the players has won

    if any(all(board[i][j] == p for j in range(k)) for i in range(k)):
        return True
    if any(all(board[j][i] == p for j in range(k)) for i in range(k)):
        return True
    if all(board[i][i] == p for i in range(k)):
        return True
    if all(board[i][-(i + 1)] == p for i in range(k)):
        return True

    return False

start_game()
```
